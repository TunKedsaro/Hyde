```python
from datetime import datetime
```

```python
def LastDate(dates):    
    # dates = ["31/3/2011","15/4/2016","1/1/2016","30/12/2015"]
    last_date = max(
        dates,
        key = lambda d: datetime.strptime(d, "%d/%m/%Y")
    )
    return last_date
```

```python
from datetime import datetime
from typing import Optional
```

```python
def parse_date_safe(d:str) -> Optional[datetime]:
    if not d or d == "-":
        return None
    try:
        return datetime.strptime(d,"%d/%m/%Y")
    except ValueError:
        return None

def get_latest_education(education_list: list) -> dict | None:
    dated_edu = []

    for edu in education_list:
        grad_date = parse_date_safe(edu.get("graduation_date"))
        dated_edu.append((grad_date, edu))

    # Sort: None dates go last
    dated_edu.sort(key=lambda x: (x[0] is None, x[0]), reverse=True)

    return dated_edu[0][1] if dated_edu else None

def infer_current_status(student_profile: dict) -> str:
    education  = student_profile.get("education", [])
    experience = student_profile.get("experience", [])

    if not education:
        return "unknown"

    latest_edu = get_latest_education(education)
    grad_date = latest_edu.get("graduation_date")

    # Rule 1: Still studying
    if not grad_date or grad_date == "-":
        return "student"

    # Rule 2 & 3: Graduated
    if experience and len(experience) > 0:
        return "professional"

    return "graduate"

def infer_current_status(student_profile: dict) -> str:
    education  = student_profile.get("education", [])
    experience = student_profile.get("experience", [])

    if not education:
        return "unknown"

    latest_edu = get_latest_education(education)
    grad_date  = latest_edu.get("graduation_date")
    
    # Rule 1: Still studying (No graduated date)
    if not grad_date or grad_date == "-":
        return "student"
    # Rule 2 : Graduated date and have some work exp
    elif experience and len(experience) > 0:
        return "professional"
    # Rule 3 : Got graduated date but have no work exp
    else:
        return "graduate"
```
